### Ranking DataFrames with rank()

```python
import pandas as pd
```

```python
sales = pd.Series([15, 32, 45, 21, 55, 15, 0],  index = ["Mo", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
```

```python
sales = pd.Series([15, 32, 45, 15, 55, 15, 0],  index = ["Mo", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
```

```python
sales
```

```python
sales.sort_values(ascending = False)
```

```python
sales.rank(ascending=False, method = "average").sort_values(ascending = True)
# rank를 통해서 해당 데이터의 등수를 지정해줄 수 있다. ascending을 통해서 작은게 1등인지 가장 큰 게 1등인지 지정가능하다.
# method에서 평균을 지정하면, 다음과 같이 5.0을 3번 반복하지 않고, 해당 평균값에 맞게 소수점 랭크를 부여해준다.
```

```python
sales.rank(ascending=False, method = "min", pct=True).sort_values()
# pct는 퍼센테이지(백분율)을 의미하고, 랭킹을 백분율로 나타내어 백분위를 알아볼 수 있게 도와준다.
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.fare.rank(method = "min", ascending = False)
```

```python
titanic["fare_rank"] = titanic.fare.rank(ascending = False, method="min")
```

```python
titanic.head()
```

```python
titanic.sort_values("fare", ascending= False)
```

```python
titanic.drop(columns = "fare_rank", inplace= True)
```

### apply(), map() and applymap()

```python
import pandas as pd
```

```python
sales = pd.read_csv("sales.csv", index_col = 0)
```

```python
sales
```

```python
sales.info()
```

```python
sales.min(axis = 0)
#axis가 0이면 인덱스 대상으로 분석
```

```python
sales.min(axis = 1)
# 1이면 열 대상 분석, 각 열중 대상값이 가장 작은 것
```

```python
def range(series):
    return series.max() - series.min()
```

```python
sales.apply(range, axis = 0)
#apply함수는 사용 시 함수의 리턴값을 데이터 프레임에 적용 가능하도록 도와준다.
```

```python
sales.apply(lambda x : x.max() - x.min(), axis=0)
# 람다 함수를 통해서도 적용이 가능하다.
```

```python
summer = pd.read_csv("summer.csv")
```

```python
summer.head()
```

```python
summer.Athlete.apply(lambda x: x[0])
# Athlete 데이터에 lambda함수를 적용해서 첫 문자를 불러온다.
```

```python
summer.Athlete.map(lambda x: x[0])
# map메서드를 통해서도 모든 데이터에 일괄적 함수를 적용해줄 수 있다.
```

```python
summer.iloc[:,1:3].applymap(lambda x: x[0])
# 다수의 열 데이터에 적용할 때는 applymap 메서드를 통해서 일괄 적용을 실시한다.
```

```python
sales.applymap(lambda x: 0.4*x-5)
```

```python
sales*0.4-5
```

### Hierarchical Indexing (MultiIndex) Intro

```python
import pandas as pd
```

```python
titanic
```

```python
titanic.swaplevel()
# swaplevel()메서드를 사용하면 인덱스의 출현 순서를 바꿔줄 수 있다.
```

### String Operations in Pandas

```python
import pandas as pd
```

```python
summer = pd.read_csv("summer.csv")
```

```python
summer.head()
```

```python
names = summer.loc[:9, "Athlete"].copy()
```

```python
names
```

```python
names.str.lower()
# 데이터 프레임에 문자열을 통제할 때는 앞에 str이라는 프로퍼티를 정의하여 통제해야 한다.
```

```python
names.str.title()
```

```python
summer.Event.str.split(" ", n = 2, expand= True)
# n파라미터는 스플릿으로 나눠지는 단어의 숫자으 공백을 최대 2개로 제한하는 것이다. 따라서 2개까지의 공백을 기준으로 나뉘게 된다.
# expand는 해당 결과를 토대로 각 단어별 열을 새로 만드는 것이다. 만약 False라면 하나의 열에 3개의 단어가 별도로 나뉘어 들어간다.
```

```python
summer[summer.Event.str.contains("100M")]
# contains라는 문자열 메서드를 통해서 100M를 포함하는 모든 데이터에 접근이 가능하다.
```
