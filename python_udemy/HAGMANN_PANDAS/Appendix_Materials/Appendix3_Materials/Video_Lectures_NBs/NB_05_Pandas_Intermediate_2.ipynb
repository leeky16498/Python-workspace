## Advanced Filtering with between(), isin() and ~

```python
import pandas as pd
```

```python
summer = pd.read_csv("summer.csv")
```

```python
summer.head()
summer.iloc[:5] = np.NaN
summer
```

```python
og_1988 = summer.loc[summer.Year == 1988]
```

```python
og_1988.head()
```

```python
og_1988.tail()
```

```python
og_1988.info()
```

```python
og_since1992 = summer.loc[summer.Year >= 1992] ## loc을 통해서 행의 조건값을 통해서 데이터를 걸러낼 수 있다.
```

```python
og_since1992.head()
```

```python
og_since1992.tail()
```

```python
summer.Year.between(1960, 1969).head() # between 을 사용하면 범위 내의 데이터 인지를 확인해서 리턴해준다.
```

```python
og_60s = summer.loc[summer.Year.between(1960, 1969, inclusive=True)]
```

```python
og_60s.head()
```

```python
og_60s.tail()
```

```python
my_favourite_games = [1972, 1996]
```

```python
summer.Year.isin(my_favourite_games).head()
```

```python
og_72_96 = summer.loc[summer.Year.isin(my_favourite_games)]
```

```python
og_72_96.head()
```

```python
og_72_96.tail()
```

```python
og_not_72_96 = summer.loc[~summer.Year.isin(my_favourite_games)]
```

```python
og_not_72_96.head()
```

```python
og_not_72_96.Year.unique()
```

## Handling NA Values / missing Values

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic.isna().sum() # 두가지 메서드가 있다. isna, notna 결측이다, 결측이 아니다 여부를 체크하는 메서드이다.
```

```python
titanic.notna().sum() # notna = 결측값이 아닌 것들의 총 갯수를 센 것이다. 
```

```python
titanic.loc[titanic.embarked.isna()] #isna메서드를 사용하면 결측값 여부에 대한 True False를 리턴한다.
```

```python
titanic.shape
```

```python
titanic.dropna() # dropna를 통해 최소 하나 이상의 결측값을 가진 행을 삭제한다.
```

```python
titanic.dropna().shape
```

```python
titanic.dropna(how = "all").shape # how를 통해 결측값 지우는 여부를 설정가능하다.
```

```python
titanic.dropna(axis = 1, how = "any").shape # axis에 1을 넣어주게 되면, 열을 삭제한다. 
```

```python
titanic.dropna(axis = 1, thresh = 500).shape # 
```

```python
titanic.dropna(axis = 1, thresh = 500, inplace = True) # 결측값의 갯수가 500개 이상인 열을 삭제한다. thresh는 한계선을 정해주는 파라미터임.
```

```python
titanic.info()
```

```python
titanic.loc[titanic.age.isna()]
```

```python
mean_age = titanic.age.mean()
mean_age
```

```python
titanic.age.fillna(value = mean_age, inplace = True) # fillna나이열에서 결측값에 대해서 평균값을 적용한다. fillna(0)는 0을 넣어라라는 의미.
```

```python
titanic.age
```

```python
titanic.info()
```
