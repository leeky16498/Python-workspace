## Part 6: Data Visualization with Matplotlib

## The plot() method

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
```

```python
titanic.plot(subplots= True, figsize=(15, 12), sharex=True, sharey=False) 
# 플롯으로 그리기 위한 파라미터에 넣어준다.
# 서브 플롯을 True로 해주면 각 열별 데이터를 그려낼 수 있다. 
# sharex는 x축 인덱스를 공유할 것인가,y도 마찬가지이다. 
plt.show() # 쇼로 보여준다. 맷플롯립은 기본적으로 인덱스를 x로 기준하여 나타내준다.
```

```python
titanic.age.plot(figsize=(12, 8)) #figsize : 나타나는 차트의 가로세로 길이를 나타낸다. 
plt.show()
```

## Customization

```python
xticks = [x for x in range(0,901, 50)]
xticks
```

```python
yticks = [y for y in range(0, 81, 5)]
yticks
```

```python
plt.style.available
```

```python
plt.style.use("classic")
```

```python
titanic.age.plot(figsize = (12,8), fontsize= 13, c = "r", linestyle = "-",
                 xlim = (0,900), ylim = (0,80), xticks = xticks, yticks = yticks, rot = 45) 
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = 3, fontsize = 15) # 그래프에 대한 범례를 어디에 나타낼 것인가에 대한 것이다. 아래 - age라는 것을 볼 수 있다. 
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
#plt.grid()
plt.show()
```

## Histograms

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.age.value_counts()
```

```python
titanic.age.plot(kind = "hist", figsize = (12,8), fontsize = 15, bins = 80, density = True)
plt.show()
```

```python
titanic.age.hist(figsize = (12,8), bins = 80, xlabelsize=15, ylabelsize= 15, cumulative = True)
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.hist(titanic.age.dropna(), bins = 80, density = True, cumulative= True)
plt.show()
```
