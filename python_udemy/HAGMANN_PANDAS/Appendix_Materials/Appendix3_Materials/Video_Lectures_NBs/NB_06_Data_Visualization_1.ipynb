## The plot() method

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
```

```python
titanic.plot()
plt.show()
```

```python
titanic.plot(subplots= True, figsize=(15, 12), sharex=True, sharey=False) 
# 플롯으로 그리기 위한 파라미터에 넣어준다.
# 서브 플롯을 True로 해주면 각 열별 데이터를 그려낼 수 있다. 
# sharex는 x축 인덱스를 공유할 것인가,y도 마찬가지이다. 
plt.show() # 쇼로 보여준다. 맷플롯립은 기본적으로 인덱스를 x로 기준하여 나타내준다.
```

```python
titanic.age.plot(figsize=(12, 8)) #figsize : 나타나는 차트의 가로세로 길이를 나타낸다. 
plt.show() # 플롯을 그리고 난 후에 이렇게 쇼를 해줘야 한다.
```

## Customization

```python
xticks = [x for x in range(0,901, 50)]
xticks
```

```python
yticks = [y for y in range(0, 81, 5)]
yticks
```

```python
plt.style.available
```

```python
plt.style.use("classic")
```

```python
titanic.age.plot(figsize = (12,8), fontsize= 13, c = "r", linestyle = "-",
                 xlim = (0,900), ylim = (0,80), xticks = xticks, yticks = yticks, rot = 45)
#figsize : 그림크기, fontsize : 레이블 폰트 크기, c:컬러이다 지금은 빨간색으로 지정함, linestyle: 콜론, 닷, 전부 가능하며, 원하는 스타일로 그릴 수 있다.
#xlim : 리밋, ylim : 리밋,  xticks = x축 간격, yticks = y축 간격, rot은 x축 인덱스의 레이블의 회전정도를 지정해주는 것이다.

plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = 3, fontsize = 15) # 그래프에 대한 범례를 어디에 나타낼 것인가에 대한 것이다. 아래 - age라는 것을 볼 수 있다. 
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid() # 그리드 함수로 영역구분이 가능하다. 
plt.show()
```

## Histograms

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head() # 위에 5개 읽어온다.
```

```python
titanic.age.value_counts(bins=10) #bins는 데이터 간격을 의미한다. 
```

```python
titanic.age.plot(kind = "hist", figsize = (12,8), fontsize = 15, bins= 80, density = True)
#density = 상대적 값을 리턴하여 그려줄 수 있다. 
plt.show()
```

```python
titanic.age.hist(figsize = (12, 8), bins=80, xlabelsize=15, ylabelsize= 15, cumulative = True)
#plot : 파라미터에 "hist"선택을 안해도 상관이 없다. hist메서드를 통해서 바로 접근이 가능하다. 
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.hist(titanic.age.dropna(), bins = 80, density = True, cumulative= True)
#plt.hist로 바로 접근하는 것은 결측값이 있는 경우 오류를 일으키므로, 결측값 제거 후 플로팅을 해줘야 한다. 
plt.show()
```

## Scatterplots

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn") # 맷플롯립 패키지를 설치하고 그래프 스타일을 설정한다.
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.plot(kind = "scatter", figsize = (15,8), x = "age", y = "fare", c = "pclass", marker = "x", s = 20, colormap= "viridis" ) 
# 산점도, 그림크기, 각 축 변량, c는 컬러 매개변수이며 "red"입력시 빨간점으로 변함
# 또한 또 다른 열을 넣어주게 되면 3차원 그림으로 변경되게 된다.
# , marker = "x"로 하면 x점으로 변경된다. s=20은 각 점 크기이다. 컬러맵은 사용할 색상들의 집합이다.
plt.show()
```
