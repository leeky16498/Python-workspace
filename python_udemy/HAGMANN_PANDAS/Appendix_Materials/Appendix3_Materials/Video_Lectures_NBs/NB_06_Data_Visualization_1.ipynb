## The plot() method

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
```

```python
titanic.plot()
plt.show()
```

```python
titanic.plot(subplots= True, figsize=(15, 12), sharex=True, sharey=False) 
#     .
#   True       . 
# sharex x   ,y . 
plt.show() #  .    x  .
```

```python
titanic.age.plot(figsize=(12, 8)) #figsize :     . 
plt.show() #        .
```

## Customization

```python
xticks = [x for x in range(0,901, 50)]
xticks
```

```python
yticks = [y for y in range(0, 81, 5)]
yticks
```

```python
plt.style.available
```

```python
plt.style.use("classic")
```

```python
titanic.age.plot(figsize = (12,8), fontsize= 13, c = "r", linestyle = "-",
                 xlim = (0,900), ylim = (0,80), xticks = xticks, yticks = yticks, rot = 45)
#figsize : , fontsize :   , c:   , linestyle: , ,  ,     .
#xlim : , ylim : ,  xticks = x , yticks = y , rot x     .

plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = 3, fontsize = 15) #        .  - age    . 
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid() #    . 
plt.show()
```

## Histograms

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head() #  5 .
```

```python
titanic.age.value_counts(bins=10) #bins   . 
```

```python
titanic.age.plot(kind = "hist", figsize = (12,8), fontsize = 15, bins= 80, density = True)
#density =      . 
plt.show()
```

```python
titanic.age.hist(figsize = (12, 8), bins=80, xlabelsize=15, ylabelsize= 15, cumulative = True)
#plot :  "hist"   . hist    . 
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.hist(titanic.age.dropna(), bins = 80, density = True, cumulative= True)
#plt.hist        ,      . 
plt.show()
```
