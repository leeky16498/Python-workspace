```python
close = pd.read_csv("close.csv", index_col="Date", parse_dates=["Date"])
close
```

```python
msft = close.MSFT.dropna().to_frame().copy()
msft
```

```python
msft.rename(columns={"MSFT" : "Price"}, inplace=True)# 칼럼의 값을 직접 지정해서 바꿔줌.
msft
```

```python
msft["P_lag1"] = msft.shift(periods=1)
msft
```

```python
msft["P_diff"] = msft.Price.sub(msft.P_lag1) # 데이터를 빼준다.
msft
```

```python
msft["P_diff2"] = msft.Price.diff(periods=1) #디프메서드를 통해서 데이터 차이 계싼이 가능
msft
```

```python
msft.P_diff.equals(msft.P_diff2) 
```

```python
msft.Price.div(msft.P_lag1)
```

```python
msft["Returns"] = msft.Price.pct_change(periods=1)# pct는 퍼센테이지 체인지를 의미한다.
```

```python
msft
```

```python
msft.drop(columns=["P_lag1", "P_diff", "P_diff2"], inplace = True) # drop메서드로 전부 삭제 가능하다.
msft
```

```python
msft.to_csv("msft.csv")
```
